```python
# Do the process of Data Augmentation by using ImageDataGenerator

from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
# Pass the data to ImageDataGenerator for preprocessing train_data

train_data_path = '/kaggle/input/mechanical-tools-dataset/train_data/train_data'
data_gen = ImageDataGenerator(zoom_range=0.2, shear_range=0.2,
                              rotation_range=20,
                              vertical_flip=True,
                              fill_mode='nearest',
                              validation_split=0.20,
                              horizontal_flip=True,
                              rescale=1/255,
                              brightness_range=(0.2,0.2))
```

```python
# /kaggle/input/mechanical-tools-dataset/train_data/train_data/Gasoline Can
```

```python
train_data = data_gen.flow_from_directory(directory=train_data_path,
                                          target_size=(224,224),batch_size=32,color_mode='rgb',
                                          class_mode='categorical',subset='training',
                                          shuffle=True)


valid_train_data = data_gen.flow_from_directory(directory=train_data_path,
                                          target_size=(224,224),batch_size=32,color_mode='rgb',
                                          class_mode='categorical',subset='validation',
                                          shuffle=False,seed=42)
```

```python
# Pass the data to ImageDataGenerator for preprocessing Mechanical Tools Image dataset

train_data_path_1 = '/kaggle/input/mechanical-tools-dataset/Mechanical Tools Image dataset/Mechanical Tools Image dataset'
data_gen = ImageDataGenerator(zoom_range=0.2, shear_range=0.2,
                              rotation_range=20,
                              vertical_flip=True,
                              fill_mode='nearest',
                              validation_split=0.20,
                              horizontal_flip=True,
                              rescale=1/255,
                              brightness_range=(0.2,0.2))
```

```python
train_data_1 = data_gen.flow_from_directory(directory=train_data_path_1,
                                          target_size=(224,224),batch_size=32,color_mode='rgb',
                                            class_mode='categorical',subset='training',
                                               shuffle=True)

valid_train_data_1 = data_gen.flow_from_directory(directory=train_data_path_1,
                                          target_size=(224,224),batch_size=32,color_mode='rgb',
                                            class_mode='categorical',subset='validation',
                                               shuffle=False,seed=42)
```

```python
# Build CNN Model by from tensorflow keras

from tensorflow.keras.models import Sequential,Model
from tensorflow.keras.layers import Conv2D,MaxPool2D,Dense,Flatten
```

```python
model = Sequential()
model.add(Conv2D(32,(3,3), input_shape=(224,224,3)))
model.add(MaxPool2D())
model.add(Conv2D(64,(3,3)))
model.add(MaxPool2D())
model.add(Conv2D(128,(3,3)))
model.add(MaxPool2D())
model.add(Conv2D(128,(3,3)))
model.add(MaxPool2D())
model.add(Conv2D(256,(3,3)))
model.add(MaxPool2D())
model.add(Flatten())
model.add(Dense(8,activation='softmax'))

model.compile(loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(train_data, epochs=10)
```

```python
model.fit(train_data_1, epochs=10)
```

```python
# Transfer Learning ---> Using InceptionV3
```

```python
from tensorflow.keras.applications.inception_v3 import InceptionV3
```

```python
inception = InceptionV3(include_top=False, input_shape=(224,224,3))
```

```python
inception.summary()
```

```python
inception.output
```

```python
final_layer = Dense(8,activation = 'softmax')(Flatten()(inception.output))
```

```python
inception_model = Model(inputs = inception.input,
                       outputs = final_layer)
inception.summary()
```

```python
inception_model.compile(loss='categorical_crossentropy',
                       metrics = ['accuracy'])
inception_model.fit_generator(train_data,epochs=40)
```

```python
inception_model.compile(loss='categorical_crossentropy',
                       metrics = ['accuracy'])
inception_model.fit_generator(train_data_1,epochs=40)
```

```python
inception_model.save('train_data_model_inception.h5')
```

```python
# Transfer Learning == VDD model
```

```python
from tensorflow.keras.applications.vgg16 import VGG16
```

```python
vgg = VGG16(include_top=False, input_shape=(224,224,3))
```

```python
vgg.summary()
```

```python
for layers in vgg.layers:
    layers.trainable = False
```

```python
vgg.summary()
```

```python
vgg.output
```

```python
final_layer = Dense(8,activation = 'softmax')(Flatten()(vgg.output))
```

```python
vgg_model = Model(inputs = vgg.input,
                       outputs = final_layer)
vgg.summary()
```

```python
vgg_model.compile(loss='categorical_crossentropy',
                       metrics = ['accuracy'])
vgg_model.fit_generator(train_data,epochs=40)
```

```python
vgg_model.compile(loss='categorical_crossentropy',
                       metrics = ['accuracy'])
vgg_model.fit_generator(train_data_1,epochs=40)
```

```python
import cv2
```

```python
def predict(path):
    img = cv2.imread(path)
    img = cv2.resize(img,(224,224))/255
    y = vgg_model.predict_on_batch(img.reshape(1,224,224,3)).argmax()
    if y == 0:
        return 'Gasoline Can'
    elif y == 1:
        return 'Hammer'
    elif y == 2:
        return 'Pilers'
    elif y == 3:
        return 'Rope'
    elif y == 4:
        return 'Screw Driver'
    elif y == 5:
        return 'Tool box'
    elif y == 6:
        return 'Wrench'
    else:
        return 'pebble'
```

```python
imgs = os.listdir('/kaggle/input/mechanical-tools-dataset/test_data/test_data')
for img in imgs:
    y = predict('/kaggle/input/mechanical-tools-dataset/test_data/test_data/'+img)
    print(img,': ',y)
```
